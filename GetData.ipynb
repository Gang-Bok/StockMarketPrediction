```python
import FinanceDataReader as fdr
from sklearn.preprocessing import StandardScaler
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["figure.figsize"] = (14,4)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["axes.grid"] = True
```

### Get S&P 500 Dataset

```python
stocks = fdr.StockListing('S&P500')
```

```python
stocks
```

### Data Preprocessing

```python
cnt = 0
stock_name = []
init_len = len(fdr.DataReader('MMM', '2017'))
for idx in stocks['Symbol']:
    dataset_iter = fdr.DataReader(idx,'2017')
    
    dataset_iter.drop(['Open','High','Low','Close','Volume'], axis = 'columns', inplace = True)
    
    dataset_iter_target = dataset_iter
    dataset_iter_sequence = dataset_iter
    
    # 신규 상장 제거
    if len(dataset_iter) < init_len:
        continue
    
    stock_name.append(idx)
    
    # 데이터 표준화 진행
    std_scaler = StandardScaler()
    std_train = std_scaler.fit_transform(dataset_iter_sequence)
    std_dataset = pd.DataFrame(std_train, columns=dataset_iter_sequence.columns,
                         index=list(dataset_iter_sequence.index.values))
    
    for i in range(1,240):
        name2 = 'Change' + str(i)
        std_dataset[name2] = std_dataset['Change'].shift(-i)
    std_dataset = std_dataset.dropna(axis=0)
    std_train_dataset = std_dataset[:-230]
    std_test_dataset = std_dataset[-230:]
    
    name3 = 'testData/s' + str(cnt) + '.csv'
    std_test_dataset.to_csv(name3)
    if idx == 'MMM':
        train_set = std_train_dataset
    else :
        train_set = pd.concat([train_set,std_train_dataset], axis = 0)
    
    name = 'Change' + str(cnt)
    dataset_iter_target.columns = [name]
    if idx == 'MMM':
        dataset_target = dataset_iter_target
    else :
        dataset_target = pd.concat([dataset_target,dataset_iter_target], axis = 1)
    print("iter %d is complete" % cnt)
    cnt+=1
```

### Making Target Dataset

```python
median = dataset_target.median(axis=1)
```

```python
median
```

```python
dataset_target = pd.concat([dataset_target, median], axis = 1)
```

```python
dataset_target = dataset_target.rename({0:'Median'}, axis='columns')
```

```python
dataset_target
```

```python
cnt = 0
for i in range(0, len(dataset_target.columns)-1):
    name  = "Change" + str(cnt)
    dataset_target.loc[dataset_target[name] < dataset_target['Median'], name] = -100
    dataset_target.loc[dataset_target[name] >= dataset_target['Median'], name] = 100
    dataset_target.loc[dataset_target[name] == 100, name] = 1
    dataset_target.loc[dataset_target[name] == -100, name] = 0
    cnt +=1
```
